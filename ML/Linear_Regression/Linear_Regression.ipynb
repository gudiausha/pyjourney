```python
# convert the columns(x and y) in both training and testing sets to matrices/numpy arrays
x_train = np.array(train_data['x'])
y_train = np.array(train_data['y'])
x_test = np.array(test_data['x'])
y_test = np.array(test_data['y'])

# reshape the x array in both training and testing set to (-1,1) otherwise error occurs when passing input to the model
x_train_shape = x_train.reshape(-1,1)
x_test_shape = x_test.reshape(-1,1)

# create an object for the linear regression model
regr = linear_model.LinearRegression(normalize=True)
# Train the model using the training sets
regr.fit(x_train_shape, y_train)
# give testing data as input and get the predicted value
y_pred = regr.predict(x_test_shape)
# prints the R2 score of the model
print('r2: ',r2_score(y_test,y_pred))
# prints the Mean Squared Error of the model
print('MSE: ',mean_squared_error(y_test,y_pred))
```
